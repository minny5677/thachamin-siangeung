 - ‡∏ñ‡πâ‡∏≤ True ‡∏à‡∏∞‡∏™‡∏£‡πâ‡∏≤‡∏á dict ‡∏ó‡∏µ‡πà‡∏°‡∏µ key ‡πÄ‡∏õ‡πá‡∏ô‡∏Ñ‡πà‡∏≤‡∏à‡∏≤‡∏Å list ‡∏ï‡∏±‡∏ß‡∏ó‡∏µ‡πà 1 ‡πÅ‡∏•‡∏∞ value ‡∏à‡∏≤‡∏Å list ‡∏ï‡∏±‡∏ß‡∏ó‡∏µ‡πà 2
 - ‡∏ñ‡πâ‡∏≤ False ‡∏à‡∏∞‡∏™‡∏£‡πâ‡∏≤‡∏á dict ‡∏ó‡∏µ‡πà‡∏°‡∏µ key ‡πÄ‡∏õ‡πá‡∏ô‡∏Ñ‡πà‡∏≤‡∏à‡∏≤‡∏Å list ‡∏ï‡∏±‡∏ß‡∏ó‡∏µ‡πà 2 ‡πÅ‡∏•‡∏∞ value ‡∏à‡∏≤‡∏Å list ‡∏ï‡∏±‡∏ß‡∏ó‡∏µ‡πà 1
 - output ‡∏Ç‡∏≠‡∏á function ‡∏Ñ‡∏∑‡∏≠ dict ‡∏ó‡∏µ‡πà‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏Ç‡∏∂‡πâ‡∏ô 
 ‡πÄ‡∏ä‡πà‡∏ô
 output1 = createDict( [ 'apple' , 'banana' ] , [ 1 , 2 ] , True )
 ‡∏à‡∏∞‡πÑ‡∏î‡πâ output1 ‡πÄ‡∏õ‡πá‡∏ô { 'apple' : 1 , 'banana' : 2 }
 output2 = createDict( [ 'apple' , 'banana' ] , [ 1 , 2 ] , False )
 ‡∏à‡∏∞‡πÑ‡∏î‡πâ output2 ‡πÄ‡∏õ‡πá‡∏ô { 1 : 'apple' , 2 : 'banana' }

```python
def createDict(x1,x2,y):
    A = {}
    for i in range(len(x1)):
        if y == True:
            A[x1[i]] = x2[i]
        else:
            A[x2[i]] = x1[i]
    return(A)
            
```

```python
createDict([ 'apple' , 'banana' ] , [ 1 , 2 ] , True)
```

```python
def createDict(l_text,l_number,logicV):
    if logicV: #logicV == Trueüòä‚úî
        retdict = {}
        for i in range(len(l_text)):
            retdict[l_text[i]] = l_number[i]
    else:
        retdict = {}
        for i in range(len(l_text)):
            retdict[l_number[i]] = l_text[i]
    return  retdict    
```

```python
createDict([ 'apple' , 'banana' ] , [ 1 , 2 ] , True)
```

```python
createDict( [ 'apple' , 'banana' ] , [ 1 , 2 ] , False )
```

## how many time A,T,G,C in the sequence?

```python
def  countChar(seq,charater):
    charcount = 0
    for i in seq:
        if i == charater:
            charcount = charcount+1
    return charcount
```

```python
countChar(DNA.iloc[1][0],'A')
```

```python
countChar('CGAACTTTAAAATCTGTGTGGCTGTCACTCGGCTGCATGCTTAGTGCACTCACGCAGTATAATTAATAA','A')
```

```python
Anum = 0
Tnum = 0
Gnum = 0
Cnum = 0
for i in range(DNA.shape[0]): 
    Anum += countChar(DNA.iloc[i][0],'A') #‡∏ô‡∏±‡∏ö‡∏ï‡∏±‡∏ß‡∏≠‡∏±‡∏Å‡∏©‡∏£‡∏ó‡∏µ‡πà‡∏Å‡∏≥‡∏´‡∏ô‡∏î
    Tnum += countChar(DNA.iloc[i][0],'T')
    Gnum += countChar(DNA.iloc[i][0],'G')
    Cnum += countChar(DNA.iloc[i][0],'C')
    
print(f'There are {Anum} As,{Tnum} As,{Gnum} As,{Cnum} As,')
```

## what is the most frequent 5-char-long subsequence

```python
def pop_st(seq):
    st_list = list()
    for i in range(len(seq)-4):
        st_list.append(seq[i:i+5]) #[i:i+5]>>>[0:5] = [0,1,2,3,4]
    return st_list
```

```python
pop_st('AAGATCTACGT')
```

```python
POPDNA = []
for i in range(DNA.shape[0]):
    POPDNA = POPDNA + pop_st(DNA.iloc[i][0])
```

```python
len(POPDNA)
```

```python
unique_5char = list(set(POPDNA)) #‡πÉ‡∏ä‡πâ‡∏Ñ‡∏≥‡∏™‡∏±‡πà‡∏á‡πÄ‡∏ã‡∏ï‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏Å‡∏≥‡∏à‡∏±‡∏î‡∏ï‡∏±‡∏ß‡∏ó‡∏µ‡πà‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏ï‡∏±‡∏ß‡∏ã‡πâ‡∏≥‡∏Å‡∏±‡∏ô‡πÄ‡∏•‡∏¢‡πÉ‡∏ô 5 ‡∏ï‡∏±‡∏ß
unique_5char
```

```python
len(unique_5char)
```

```python
dic_5char = {}
for sub_str in unique_5char: #‡∏ß‡∏ô1022‡∏£‡∏≠‡∏ö
    num = 0
    for s in POPDNA: #‡∏ß‡∏ô 28191
        if sub_str == s:
            num += 1
    dic_5char[sub_str] = num
```

```python
dic_5char
```

```python
dic_5char['TGGTG']
```

```python
maxVal = -100
maxKey = None
for k in dic_5char.keys():
    if dic_5char[k] > maxVal:
        maxVal = dic_5char[k]
        maxKey = k
```

```python
print(f'max 5-long-sequence is {maxKey} with frequency equals {maxVal}')
```
