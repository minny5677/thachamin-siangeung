```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = 'C://Users//Minny//Downloads//2019_nCoV_data.csv'
covid= pd.read_csv(data)
```

```python
covid
```

```python
covid['Province/State']=covid['Province/State'].fillna('know')
```

```python
covid_china = covid[covid['Country']=='Mainland China']
```

```python
covid_china
```

```python
covid_china_group = covid_china.groupby('Province/State').max().reset_index()
```

```python
covid_china_group2= covid_china_group[covid_china_group['Province/State']!='Hubei']
```

```python
plt.bar(covid_china_group2.iloc[:,0],covid_china_group2.iloc[:,5],color = 'g')
plt.title(u'จำนวนผู้ติดเชื้อในแต่ละเมือง',fontname='Tahoma',fontsize='13')
plt.xlabel('Province/State',fontname='Tahoma',fontsize='11')
plt.ylabel('Confirmed')
```

```python
plt.bar(covid_china_group2.iloc[:,0],covid_china_group2.iloc[:,6],color = 'r')
plt.title(u'จำนวนผู้ติดเชื้อที่เสียชีวิตในแต่ละเมือง',fontname='Tahoma',fontsize='13')
plt.xlabel('Province/State',fontname='Tahoma',fontsize='11')
plt.ylabel('Deaths')
```

```python
plt.bar(covid_china_group2.iloc[:,0],covid_china_group2.iloc[:,7],color = 'b')
plt.title(u'จำนวนผู้ติดเชื้อที่รักษาหายในแต่ละเมือง',fontname='Tahoma',fontsize='13')
plt.xlabel('Province/State',fontname='Tahoma',fontsize='11')
plt.ylabel('Recovered')
```
