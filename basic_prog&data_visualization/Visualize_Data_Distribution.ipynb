# Scatter plot

```python
irisdata =pd.read_csv('C:\\Users\\Admin\\Downloads\\iris.data',header=None)
irisdata
```

```python
setosa_np = irisdata.iloc[:50,:-1].to_numpy()
setosa_np[:5,:]
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1]) #ให้ x=setosa_np[:,0] คือเอาทุกแถว,หลักที่ 0 ให้  y=setosa_np[:,1]คือ เอาทุกแถว,หลักที่ 1
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=6)
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=10*setosa_np[:,2]) # s=10*setosa_np[:,2] คือ การปรับขนาดโดยเอา 10*เข้าไปในsetosa_np[:,2]
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=10*setosa_np[:,2], c ='k') # c ='k'คือ การใส่สี
```

```python
print(f'max = {max(setosa_np[:,3])},min = {min(setosa_np[:,3])}')
```

```python
colors = []
for i in setosa_np[:,3]:
    if i<0.15:
        colors.append('b')
    elif i<0.3:
        colors.append('g')
    elif i<0.45:
        colors.append('y')
    else:
        colors.append('r')
colors
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2], c =colors)
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=10*setosa_np[:,2], c =np.array([[i,0.5,1]]))
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=20*setosa_np[:,2], c =np.array([[1,0.5,1]]))
```

```python
versicolor_np = irisdata.iloc[50:100,:-1].to_numpy()
```

```python
plt.scatter(setosa_np[:,0], setosa_np[:,1], s = 20*setosa_np[:,2], c=setosa_np[:,3],cmap = 'YlGn',marker='^')
plt.scatter(versicolor_np[:,0], versicolor_np[:,1], s = 20*versicolor_np[:,2], c=versicolor_np[:,3],cmap = 'YlGn',marker='x')
plt.scatter(versicolor_np[:,0], versicolor_np[:,1], s = 20*versicolor_np[:,2], c=versicolor_np[:,3],cmap = 'YlGn',marker='*')
plt.show()
```
